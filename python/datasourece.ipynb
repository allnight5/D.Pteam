```python
def Frequency(i):
    mtem3 = np.nanpercentile(df['최저기온'][912*i:912*(i+1)], 5,axis=None)
    mtem2 = np.nanpercentile(df['최저기온'][912*i:912*(i+1)], 25,axis=None) 
    mtem1 = np.nanpercentile(df['최저기온'][912*i:912*(i+1)], 60,axis=None)  
    mtem1 = float(round(mtem1, 2))
    mtem2 = float(round(mtem2, 2))
    mtem3 = float(round(mtem3, 2))
    
    pha1 = np.nanpercentile(df['기압'][912*i:912*(i+1)], 95,axis=None) 
    pha2 = np.nanpercentile(df['기압'][912*i:912*(i+1)], 75,axis=None) 
    pha3 = np.nanpercentile(df['기압'][912*i:912*(i+1)], 40,axis=None)  
    pha1 = float(round(pha1, 2))
    pha2 = float(round(pha2, 2))
    pha3 = float(round(pha3, 2))
    
    Oz1 = np.nanpercentile(df['오존량'][912*i:912*(i+1)], 95,axis=None)
    Oz2 = np.nanpercentile(df['오존량'][912*i:912*(i+1)], 75,axis=None)
    Oz3 = np.nanpercentile(df['오존량'][912*i:912*(i+1)], 40,axis=None) 
    Oz1 = float(round(Oz1, 2))
    Oz2 = float(round(Oz2, 2))
    Oz3 = float(round(Oz3, 2)) 
    
    ra1 = np.nanpercentile(df['일교차'][912*i:912*(i+1)], 95,axis=None)
    ra2 = np.nanpercentile(df['일교차'][912*i:912*(i+1)], 75,axis=None)
    ra3 = np.nanpercentile(df['일교차'][912*i:912*(i+1)], 40,axis=None) 
    ra1 = float(round(ra1, 2))
    ra2 = float(round(ra2, 2))
    ra3 = float(round(ra3, 2))
    
    hum1 = np.nanpercentile(df['최소상대습도'][912*i:912*(i+1)], 5,axis=None)
    hum2 = np.nanpercentile(df['최소상대습도'][912*i:912*(i+1)], 25,axis=None)
    hum3 = np.nanpercentile(df['최소상대습도'][912*i:912*(i+1)], 60,axis=None) 
    hum1 = float(round(hum1, 2))
    hum2 = float(round(hum2, 2))
    hum3 = float(round(hum3, 2))
    ALL = mtem1, mtem2, mtem3, pha1, pha2, pha3, Oz1, Oz2, Oz3, ra1, ra2, ra3, hum1, hum2, hum3 
    return ALL
```

```python
def weight(i):
    temr = df['일교차'][912*i:912*(i+1)].corr(df['환자수'][912*i:912*(i+1)])
    hum = df['최소상대습도'][912*i:912*(i+1)].corr(df['환자수'][912*i:912*(i+1)])
    oj = df['오존량'][912*i:912*(i+1)].corr(df['환자수'][912*i:912*(i+1)])
    atm = df['기압'][912*i:912*(i+1)].corr(df['환자수'][912*i:912*(i+1)])
    mtem = -df['최저기온'][912*i:912*(i+1)].corr(df['환자수'][912*i:912*(i+1)])

    al = temr+hum+oj+mtem+atm 

    temr,hum,oj,mtem,atm
    temr = temr/al
    hum = hum/al
    oj = oj/al
    atm = atm/al
    mtem = mtem/al
    return temr, mtem, hum, atm, oj
```
