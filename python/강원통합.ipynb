```python
for i in 강원통합2.index:
    if 강원통합2.values[i,2] < 17 :
        if 강원통합2.values[i,6] >70:
            강원통합2['습도위험지수'].values[i] = '1'
        elif 59< 강원통합2.values[i,6] <70:
            강원통합2['습도위험지수'].values[i] = '2'
        elif 49< 강원통합2.values[i,6] <60:
            강원통합2['습도위험지수'].values[i] = '3'
        else:
            강원통합2['습도위험지수'].values[i] = '4'
    elif 17< 강원통합2.values[i,2] < 21:
        if 강원통합2.values[i,6] >60:
            강원통합2['습도위험지수'].values[i] = '1'
        elif 49< 강원통합2.values[i,6] <60:
            강원통합2['습도위험지수'].values[i] = '2'
        elif 39< 강원통합2.values[i,6] <50:
            강원통합2['습도위험지수'].values[i] = '3'
        else:
            강원통합2['습도위험지수'].values[i] = '4'
            
    elif 20<강원통합2.values[i,2] < 24:
        if 강원통합2.values[i,6] >50:
            강원통합2['습도위험지수'].values[i] = '1'
        elif 39< 강원통합2.values[i,6] <50:
            강원통합2['습도위험지수'].values[i] = '2'
        elif 29< 강원통합2.values[i,6] <40:
            강원통합2['습도위험지수'].values[i] = '3'
        else:
            강원통합2['습도위험지수'].values[i] = '4'
    elif 강원통합2.values[i,2] > 23:
        if 강원통합2.values[i,6] >40:
            강원통합2['습도위험지수'].values[i] = '1'
        elif 29< 강원통합2.values[i,6] <40:
            강원통합2['습도위험지수'].values[i] = '2'
        elif 19< 강원통합2.values[i,6] <30:
            강원통합2['습도위험지수'].values[i] = '3'
        else:
            강원통합2['습도위험지수'].values[i] = '4'
```

```python
강원통합2['일교차'] = 강원통합2['최고기온(°C)'] -  강원통합2['최저기온(°C)']
```

```python
for i in 강원통합2.index:    
    if 강원통합2.values[i,3] <6 or 29<강원통합2.values[1,2]:
        강원통합2['최저기온위험지수'].values[i] = '4' 
    elif 5<강원통합2.values[i,4] <15:
        강원통합2['최저기온위험지수'].values[i] = '3' 
    elif 14<강원통합2.values[i,4] <25:
        강원통합2['최저기온위험지수'].values[i] = '2'  
    elif 24< 강원통합2.values[i,4] <30:
        강원통합2['최저기온위험지수'].values[i] = '1' 
        
    if 강원통합2.values[i,9] < 3 :  
        강원통합2['일교차위험지수'].values[i] = '1' 
    elif 2< 강원통합2.values[i,12] < 6: 
        강원통합2['일교차위험지수'].values[i] = '2'  
    elif 5<강원통합2.values[i,12] < 10:  
        강원통합2['일교차위험지수'].values[i] = '3'  
    elif 9< 강원통합2.values[i,12] : 
        강원통합2['일교차위험지수'].values[i] = '4' 
```
