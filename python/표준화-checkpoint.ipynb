```python
chong2['최저기온표준화']=(chong2['최저기온(°C)']-chong2['최저기온(°C)'].mean())/chong2['최저기온(°C)'].std()
```

```python
chong2
```

```python
chong2['일교차표준화']=(chong2['일교차']-chong2['일교차'].mean())/chong2['일교차'].std()
```

```python
chong2['습도표준화']=(chong2['평균 상대습도(%)']-chong2['평균 상대습도(%)'].mean())/chong2['평균 상대습도(%)'].std()
```

```python
chong2['오존표준화']=(chong2['평균 오존전량(DU)']-chong2['평균 오존전량(DU)'].mean())/chong2['평균 오존전량(DU)'].std()
```

```python
chong2['발생건수표준화']=(chong2['발생건수(건)']-chong2['발생건수(건)'].mean())/chong2['발생건수(건)'].std()
```
