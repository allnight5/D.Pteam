```python
for i in all_T.index:
    if all_T.values[i,2] < 17 :
        if all_T.values[i,5] >70:
            all_T['습도위험지수'].values[i] = '1'
        elif 59< all_T.values[i,5] <70:
            all_T['습도위험지수'].values[i] = '2'
        elif 49< all_T.values[i,5] <60:
            all_T['습도위험지수'].values[i] = '3'
        else:
            all_T['습도위험지수'].values[i] = '4'
            
    elif 17<all_T.values[i,2] < 21:
        if all_T.values[i,5] >60:
            all_T['습도위험지수'].values[i] = '1'
        elif 49< all_T.values[i,5] <60:
            all_T['습도위험지수'].values[i] = '2'
        elif 39< all_T.values[i,5] <50:
            all_T['습도위험지수'].values[i] = '3'
        else:
            all_T['습도위험지수'].values[i] = '4'
            
    elif 20<all_T.values[i,2] < 24:
        if all_T.values[i,5] >50:
            all_T['습도위험지수'].values[i] = '1'
        elif 39< all_T.values[i,5] <50:
            all_T['습도위험지수'].values[i] = '2'
        elif 29< all_T.values[i,5] <40:
            all_T['습도위험지수'].values[i] = '3'
        else:
            all_T['습도위험지수'].values[i] = '4' 
            
    elif all_T.values[i,2] > 23:
        if all_T.values[i,5] >40:
            all_T['습도위험지수'].values[i] = '1'
        elif 29< all_T.values[i,5] <40:
            all_T['습도위험지수'].values[i] = '2'
        elif 19< all_T.values[i,5] <30:
            all_T['습도위험지수'].values[i] = '3'
        else:
            all_T['습도위험지수'].values[i] = '4'
    
all_T.head()
```

```python
all_T['일교차'] = all_T['최고기온(°C)'] - all_T['최저기온(°C)']
```

```python
for i in all_T.index:
    if all_T.values[i,3] <-4.8 or 30<=all_T.values[1,2]:
        all_T['최저기온위험지수'].values[i] = '4' 
    elif -4.8<=all_T.values[i,3] <1.6:
        all_T['최저기온위험지수'].values[i] = '3' 
    elif 1.6<=all_T.values[i,3] <13.8:
        all_T['최저기온위험지수'].values[i] = '2'  
    elif 13.8< all_T.values[i,3] <30:
        all_T['최저기온위험지수'].values[i] = '1'  
        
    if all_T.values[i,9] < 8.7 : 
        all_T['일교차위험지수'].values[i] = '1' 
    elif 8.7<= all_T.values[i,9] < 12: 
        all_T['일교차위험지수'].values[i] = '2' 
    elif 12<=all_T.values[i,9] < 15.5:  
        all_T['일교차위험지수'].values[i] = '3' 
    elif 15.5<= all_T.values[i,9] : 
        all_T['일교차위험지수'].values[i] = '4' 

all_T.head()
```

```python
plt.figure(figsize=(15,6))
plt.plot(all_T['일교차위험지수'], color = 'green')
plt.xlabel('일시')
plt.ylabel('일교차위험지수')
plt.grid()
plt.show()
```
