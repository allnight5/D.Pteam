```python
import matplotlib.pyplot as plt
```

```python
colormap=asd['일교차표준화']
plt.figure(figsize=(30,10))
plt.scatter(asd['일시'],asd['일교차표준화'],s=20,c=colormap,marker='^')
plt.plot(asd['일시'],asd['발생건수표준화'], color = 'green')
plt.xticks([0, 100, 200, 300, 400, 500, 600, 700])
plt.colorbar()
plt.xlabel('날짜')
plt.ylabel('일교차표준화')
plt.grid()
plt.show()
```
