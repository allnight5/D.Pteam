```python
kor_tem = pd.read_csv('../Date2/six.csv' ,thousands=',', encoding='euc-kr')
```

```python
kor_tem.head()
```

```python
for i in kor_tem.index:
    if kor_tem.values[i,2] < 17 :
        if kor_tem.values[i,5] >70:
            kor_tem['습도위험지수'].values[i] = '1'
        elif 59< kor_tem.values[i,5] <70:
            kor_tem['습도위험지수'].values[i] = '2'
        elif 49< kor_tem.values[i,5] <60:
            kor_tem['습도위험지수'].values[i] = '3'
        else:
            kor_tem['습도위험지수'].values[i] = '4'
    elif 17<kor_tem.values[i,2] < 21:
        if kor_tem.values[i,5] >60:
            kor_tem['습도위험지수'].values[i] = '1'
        elif 49< kor_tem.values[i,5] <60:
            kor_tem['습도위험지수'].values[i] = '2'
        elif 39< kor_tem.values[i,5] <50:
            kor_tem['습도위험지수'].values[i] = '3'
        else:
            kor_tem['습도위험지수'].values[i] = '4'
            
    elif 20<kor_tem.values[i,2] < 24:
        if kor_tem.values[i,5] >50:
            kor_tem['습도위험지수'].values[i] = '1'
        elif 39< kor_tem.values[i,5] <50:
            kor_tem['습도위험지수'].values[i] = '2'
        elif 29< kor_tem.values[i,5] <40:
            kor_tem['습도위험지수'].values[i] = '3'
        else:
            kor_tem['습도위험지수'].values[i] = '4'
    elif kor_tem.values[i,2] > 23:
        if kor_tem.values[i,5] >40:
            kor_tem['습도위험지수'].values[i] = '1'
        elif 29< kor_tem.values[i,5] <40:
            kor_tem['습도위험지수'].values[i] = '2'
        elif 19< kor_tem.values[i,5] <30:
            kor_tem['습도위험지수'].values[i] = '3'
        else:
            kor_tem['습도위험지수'].values[i] = '4'
```

```python
kor_tem.head()
```

```python
kor_tem['일교차'] = kor_tem['최고기온(°C)'] -  kor_tem['최저기온(°C)']
```

```python
kor_tem.head()
```

```python
for i in kor_tem.index:    
    if kor_tem.values[i,3] <6 or 29<kor_tem.values[1,2]:
        kor_tem['최저기온위험지수'].values[i] = '4' 
    elif 5<kor_tem.values[i,3] <15:
        kor_tem['최저기온위험지수'].values[i] = '3' 
    elif 14<kor_tem.values[i,3] <25:
        kor_tem['최저기온위험지수'].values[i] = '2'  
    elif 24< kor_tem.values[i,3] <30:
        kor_tem['최저기온위험지수'].values[i] = '1' 
        
    if kor_tem.values[i,9] < 3 :  
        kor_tem['일교차위험지수'].values[i] = '1' 
    elif 2< kor_tem.values[i,9] < 6: 
        kor_tem['일교차위험지수'].values[i] = '2'  
    elif 5<kor_tem.values[i,9] < 10:  
        kor_tem['일교차위험지수'].values[i] = '3'  
    elif 9< kor_tem.values[i,9] : 
        kor_tem['일교차위험지수'].values[i] = '4'  
```
